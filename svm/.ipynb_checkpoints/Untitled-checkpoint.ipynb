```python
def debug(variable):
    print variable, '=', repr(eval(variable))

def findTallest(number_of_students, students, tallest_index = 0):
    print('tallest_index: {}'.format(tallest_index))
    tallest_height = students[tallest_index]
    tallest_can_see = 0
    
    next_tallest_height = 0 
    next_tallest_can_see = 0
    for index in range(tallest_index+1, number_of_students):
        tallest_can_see = tallest_can_see + 1
        next_tallest_height = students[index]
        if next_tallest_height >= tallest_height:
            print('a55555555555')
            print('next_tallest_height: {}'.format(next_tallest_height))
            print('tallest_height: {}'.format(tallest_height))
            print('a55555555555eeeeeee')
            next_tallest_index = index
            next_tallest_index, next_tallest_can_see = findTallest(number_of_students, students, next_tallest_index)
            
            print('sssssssssssssssssssss')
            print('tallest_can_see: {}'.format(tallest_can_see))
            print('tallest_index: {}'.format(tallest_index))
            print('bbbbbbbbbbbbbbbbbbbbbb')
    
    if next_tallest_can_see >= tallest_can_see:
        print('a6666666666666: updating')
        tallest_can_see = next_tallest_can_see
        tallest_index = next_tallest_index
        print('tallest_can_see: {}'.format(tallest_can_see))
        print('tallest_index: {}'.format(tallest_index))
        print('a6666666666666eeeeeeeee')
    
    return tallest_index, tallest_can_see
    
testcases = int('5')
testcases_input = {
    0 : ['5', '5 4 3 1 2'],
    1 : ['6', '3 2 1 5 4 1'],
    2 : ['7', '3 2 1 5 4 1 2'],
    3 : ['7', '5 2 1 5 4 1 2'],
    4 : ['5', '5 4 3 1 2']
}


for testcase in range(testcases):
    print('bbbbbbbbbbbbbbbbbbbbbbbbbbbb')
    number_of_students = int(testcases_input[testcase][0])
    students = list(map(int, testcases_input[testcase][1].split()))
    debug('number_of_students')
    debug('students')
    print('start calc')
    
    tallest_index, _ = findTallest(number_of_students, students)
    
    print(tallest_index+1)
    print('\n')
  
```
